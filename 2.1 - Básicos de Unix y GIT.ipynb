## Generar la URL para poder acceder a Github desde Colaboratory

* Asegúrate de que el archivo **credenciales.json** esté completo y se encuentre en la carpeta actual de trabajo (SemanaTec).

````python
# Vamos a importar el archivo y leer los contenidos.
import json

# Lectura del archivo
with open('credenciales.json', 'r') as myfile:
    data = myfile.read()

# Leer el formato del archivo
obj = json.loads(data)

# Vamos a guardar los datos en estas variables
GIT_USERNAME = obj['user']

# token
GIT_TOKEN = obj['token']

# Repo
GIT_REPO = obj['repo']

# Creamos la ruta al repositorio de nuestra cuenta
GIT_PATH = "https://" + GIT_USERNAME + ":" + GIT_TOKEN + "@github.com/" +\
            GIT_USERNAME + "/" + GIT_REPO + ".git"
````

## Agregar un remote nuevo a git

Veamos cuál es el remote asociado al repositorio original que clonamos.

````python
# Primero veamos a qué repositorio apunta la carpeta que acabamos de descargar
!git remote -v
````

````python
# Agreguemos el remote con set-url con la dirección de nuestro github personal
!git remote add origin "{GIT_PATH}"
````
