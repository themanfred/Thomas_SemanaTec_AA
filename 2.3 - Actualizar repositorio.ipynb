## Configurar datos de usuario

* Para que git nos deje hacer cambios en el repositorio en línea, hay que configurar el nombre de usuario y el correo.

* Desafortunadamente, cuando trabajamos en Colab debemos hacer esta configuración cada vez que trabajemos en una sesión nueva.

```
# Configuremos el usuario. Sustituye los valores.
!git config --global user.name "<TU NOMBRE>"
```

```
# Configuremos el correo
!git config --global user.email "<TU CORREO>"
```

## Agregar archivos al staging area.

* `git add` agrega archivos al área de trabajo. 

* Para agregar todos los archivos del directorio actual usamos `git add *`.

```
# Después de que hayas realizado los cambios a los notebooks, agregalos al índice y súbelos a tu repo.
!git add *
```

## Hacer commit a los cambios

* `git commit -m "mensaje"` es para hacer los cambios *"oficiales"*. Es decir, los preparamos para que se vean reflejados en el repositorio remoto.

* `git status` nos entrega un resumen de cuáles cambios están pendientes de realizarse.

```
# Commit con mensaje
!git commit -m "<MENSAJE REFERENTE A LOS CAMBIOS>"
```

```
# Revisa  el estatus
!git status
```

## Actualizar repositorio remoto

* `git push <remote> <branch>` es para enviar los cambios al repositorio y rama que especifiquemos.

* `git push <remote> --force` lo usamos cuando estamos completamente seguros que los cambios que tenemos localmente son los correctos. **Cuidado**: puede causar pérdidas de archivos.

```
# Sube los cambios. Si estás trabajando en una rama (branch) no olivides agregarla en la instrucción
!git push origin main
```
